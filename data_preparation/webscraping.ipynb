```python
URL = 'https://places.googleapis.com/v1/places:searchText'
field_mask = "places.displayName.text,places.formattedAddress,places.location"
headers = {
  'Accept': 'application/json',
  'Content-Type': 'application/json',
  "X-Goog-FieldMask": field_mask
}
```

```python
# Try Scraping
i = 0
j = 0
li_temp = []

for r in map_search_key['organization_final']:

  # call API
  search_key = r
  payload = {
      "textQuery": search_key,
      "maxResultCount": 1,
      "languageCode": "th",
      "regionCode": "TH"
  }

  res = authed_session.post(URL, headers=headers, json=payload)
  if res.status_code == 200:
    try:
      data = res.json()['places'][0]

      # create list and at it to list
      now = [r , data['displayName']['text'], data['formattedAddress'], data['location']['latitude'], data['location']['longitude']]
      li_temp.append(now)

      # count before sleep
      i = i+1
      if i == 50:
        i = 0
        j = j+1

        li.extend(li_temp)
        li_temp = []

        time.sleep(10)

      # if j == 1:
      #   break;
    except (KeyError, ValueError) as e:
      print(f'Error at row {j*100 +i} with {r}:')
      print(e)
      li_temp.append([r, 'Not Found', 'Not Found', 'Not Found', 'Not Found'])

      i = i+1
      if i == 50:
        i = 0
        j = j+1

        li.extend(li_temp)
        li_temp = []

        time.sleep(10)

      continue;


  else:
    print(f'Error at row {j*100 +i}: with {r}:')
    print(res.json())
    break;

# out of loop
li.extend(li_temp)

```
